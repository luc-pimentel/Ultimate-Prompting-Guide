```python
import json
from langchain.tools import BraveSearch

api_key = "YOUR_BRAVE_API_KEY"

def search_on_youtube(search_term):

        brave_search = BraveSearch.from_api_key(api_key= api_key)
        
        search_results = brave_search.run(search_term+" site:youtube.com")
        
        return json.loads(search_results)
```

```python
search_results = search_on_youtube('ChatGPT')
```

```python
def get_video_id(url):
    video_id = url.split('?v=')[1]
    return video_id
```

```python
from youtube_transcript_api import YouTubeTranscriptApi

def get_video_transcript(video_id: str) -> str:
    """
    Retrieves the transcript of a YouTube video given its video ID.

    Parameters:
    video_id (str): The ID of the YouTube video.

    Returns:
    str: The transcript of the YouTube video as a single string.
    """
    transcript_dicts = YouTubeTranscriptApi.get_transcript(video_id)
    transcript = ' '.join([dictionary['text'] for dictionary in transcript_dicts])
    return transcript
```

```python
video_url = search_results[0]['link']

video_id = get_video_id(video_url)

transcript = get_video_transcript(video_id)

print(transcript)
```

```python
summarizer_prompt = '''You are video transcript summarizer GPT.
I will give you a video transcript and you will create a summary for me.

Abide to the following rules:
- The summary should be short and to the point.
- All of the key points made in the video have to be on the final summary.
- Refrain from giving your personal opinions.
I only want to see what the author meant to say, not what you understood.

Answer in bullet points format and add a line break \n after each bullet.

Begin!
'''
```

```python
from langchain.chains import LLMChain
from langchain.chat_models import ChatOpenAI
from langchain.prompts import HumanMessagePromptTemplate, ChatPromptTemplate, SystemMessagePromptTemplate

sys_msg_prompt = SystemMessagePromptTemplate.from_template(summarizer_prompt)

human_msg_prompt = HumanMessagePromptTemplate.from_template('{input}')

chat_prompt = ChatPromptTemplate.from_messages([sys_msg_prompt, human_msg_prompt])


llm = ChatOpenAI(openai_api_key = "YOUR_OPENAI_API_KEY", model = 'gpt-3.5-turbo-16k')
summarizer_chain = LLMChain(llm = llm, prompt = chat_prompt)

```

```python
summarizer_chain.run(transcript)
```
